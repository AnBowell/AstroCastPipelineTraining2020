# Create mosaic of tiles that make kenya
#### This will done every band (4 each) and QA file (4 each)
#### 1) Group tiles per band and QA files
#### 2) Load each group, merge and save

#### The merge function takes one parameter, Group list of tiles to merge

<img src='./img/glcf_modistiles_sep06_crop.jpg' width='500'>

```python
path_merge = current_dir+'/MergedTiffData/' #Path to merged files

# Remove created files to avoid conflicts
for m in glob.glob(path_merge + 'tif'):
    os.remove(m)
    
# Group the tiles
for a in np.arange(2020001, 2020011):
    Band1_Reflectance  = glob.glob(path_tiff+f'/*A{a}*Reflectance*Band1*tif') # Group all band 1s
    Band2_Reflectance  = glob.glob(path_tiff+f'/*A{a}*Reflectance*Band2*tif') # Group all band 2s
    QA_Band1           = glob.glob(path_tiff+f'/*A{a}*Quality*Band1*tif') # Group all QA band 1s
    QA_Band2           = glob.glob(path_tiff+f'/*A{a}*Quality*Band2*tif') # Group all QA band 2s

    # Create a list of Group tiles
    files_to_merge = [Band1_Reflectance, Band2_Reflectance, QA_Band1, QA_Band2]

    # Loop through the list and merge
    for Tiles in files_to_merge:
        MosaicTiles.Merge(ToMerge=Tiles, file_dst=path_merge)
    print(f'A{a} Done')
```

# Extract NDVI files from counties using shape files
#### Here the NDVI function will for each merged reflectance image file:
#### 1) Use county shapefiles to extrat pixels from our ROIs
#### 2) Mask out clouds and the less qualitiy pixel using QA bands as mask
#### 3) Derive NDVI arrays over extracted ROIs
#### 4) Aggregate pixels over ROIs and save as time series

```python
# Get all Bands Files
NDVIArr_ls = []
for a in np.arange(2020001, 2020011):
    RedImg = glob.glob(path_merge+f'/*A{a}*Ref*Band1*tif')
    NIRImg = glob.glob(path_merge+f'/*A{a}*Ref*Band2*tif')
    QRedImg = glob.glob(path_merge+f'/*A{a}*Qual*Band1*tif')
    QNIRImg = glob.glob(path_merge+f'/*A{a}*Qual*Band2*tif')


    # Get NDVI files
    NDVIArr_ls.append(ExtractNDVI.NDVIProduction(RedImage=RedImg[0], NIRImage=NIRImg[0], REDmask=QRedImg[0], NIRmask=QNIRImg[0]))
```

# Create NDVI DataFrame

```python
# Create a DataFrame of the generated NDVI 
NDVI_df = pd.DataFrame()
NDVI_df_ls = []
for i in np.arange(0, 10):
    NDVI_df_ls.append(pd.DataFrame({'CountyID':NDVIArr_ls[i][0],
                                    'County':NDVIArr_ls[i][1], 
                                     'Date':NDVIArr_ls[i][2], 
                                     'Mean_NDVI': NDVIArr_ls[i][3],
                                     'Std_NDVI': NDVIArr_ls[i][4], 
                                     'Pixel_Count':NDVIArr_ls[i][5]}))

NDVI_df = pd.concat(NDVI_df_ls)

NDVI_df.head(20)
```

# Calculate VCI for Each County

```python
vci_ls = []
VCI_df = pd.DataFrame()
for c in np.arange(0, 48):
    df = NDVI_df[NDVI_df.CountyID==c]
    vci = 100*(df.Mean_NDVI-df.Mean_NDVI.min())/(df.Mean_NDVI.max()-df.Mean_NDVI.min())
    vci_ls.append(pd.DataFrame({'CountyID':c, 'Date':df.Date.values, 'VCI':vci}))
VCI_df = pd.concat(vci_ls)
```
